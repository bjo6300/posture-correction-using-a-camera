```python
# 주한코드

import tensorflow as tf
import keras.models
from keras.utils import np_utils
from keras.models import Sequential
from PIL import Image
import os, glob, numpy as np
from keras.preprocessing.image import img_to_array

image_w= 64
image_h= 64

# normal
file_link="/content/drive/MyDrive/colab/data/raw_photo/train_samepercent/front_normal_10629.jpg"

# text neck
# file_link="/content/drive/MyDrive/colab/data/raw_photo/train_samepercent/front_turtle_12979.jpg"


img = Image.open(file_link)
img = img.convert("RGB") #컬러변환
img = img.resize((image_w, image_h))
display(img)

# 모델 open
# model = tf.keras.models.load_model('/content/drive/MyDrive/colab/saved_model')
model = tf.keras.models.load_model('/content/drive/MyDrive/saved_model')

data = img_to_array(img) # 배열화
data = np.expand_dims(data,axis=0)

print(data.shape)

single_test = model.predict(data)
print(single_test)
```
